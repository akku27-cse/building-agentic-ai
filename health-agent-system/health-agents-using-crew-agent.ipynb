```python
from crewai import Agent, Task, Crew

recipe_scout = Agent(
    role="Recipe Scout",
    goal="Find recipes that stabilize blood sugar",
    backstory="You are a nutrition research expert.",
    model="groq/llama3-70b-8192",  
    temperature=0.4,
    verbose=True
)

clinical_reviewer = Agent(
    role="Clinical Reviewer",
    goal="Clinically review recipes for glycemic safety",
    backstory="You are a clinical nutritionist.",
    model="groq/llama3-70b-8192",   
    temperature=0.3,
    verbose=True
)
```

```python
from crewai import Task

scout_task = Task(
    description="List 10 low glycemic index recipes for blood sugar stability.",
    expected_output="A numbered list of 10 recipes with brief explanations.",
    agent=recipe_scout
)

review_task = Task(
    description="Review the recipes and rank the top 3 safest options.",
    expected_output="A ranked list of top 3 recipes with clinical justification and caution notes.",
    agent=clinical_reviewer
)
```

```python
crew = Crew(
    agents=[recipe_scout, clinical_reviewer],
    tasks=[scout_task, review_task],
    verbose=True
)

result = crew.kickoff()
print(result)
```

```python
from crewai import Agent, Task, Crew

recipe_scout = Agent(
    role="Recipe Scout",
    goal="Find recipes that stabilize blood sugar",
    backstory="You are a nutrition research expert.",
    model="groq/llama3-70b-8192",  
    temperature=0.4,
    verbose=True
)

clinical_reviewer = Agent(
    role="Clinical Reviewer",
    goal="Clinically review recipes for glycemic safety",
    backstory="You are a clinical nutritionist.",
    model="groq/llama3-70b-8192",   
    temperature=0.3,
    verbose=True
)
# --- Email Agent ---
# --- Email Agent ---
def send_email(final_result):
    email_user = os.environ.get("EMAIL_USER")
    email_pass = os.environ.get("EMAIL_APP_PASSWORD")
    recipient_email = "janaa7223@gmail.com"  # or any other email

    # Summarize the result in a professional email
    subject = "Blood Sugar Stabilizing Recipes - Review Summary"
    body = f"""
Hello,

Here is the summarized review of recipes for blood sugar stability:

{final_result}

Best regards,
Automated Recipe Review System
"""

    msg = MIMEMultipart()
    msg['From'] = email_user
    msg['To'] = "janaa7223@gmail.com"
    msg['Subject'] = subject
    msg.attach(MIMEText(body, 'plain'))

    # Send email via Gmail SMTP
    with smtplib.SMTP_SSL("smtp.gmail.com", 465) as server:
        server.login(email_user, email_pass)
        server.send_message(msg)

    return "Email sent successfully."

# Wrap it into an Agent for Crew
email_agent = Agent(
    role="Email Agent",
    goal="Summarize final review and send via Gmail",
    backstory="You take the final review result, make it into a professional email, and send it to the designated recipient using environment credentials.",
    model="groq/llama3-70b-8192",
    temperature=0.3,
    verbose=True,
    # Add a custom run function to execute sending
    run=lambda final_result: send_email(final_result)
)
```

```python
from crewai import Task

scout_task = Task(
    description="List 10 low glycemic index recipes for blood sugar stability.",
    expected_output="A numbered list of 10 recipes with brief explanations.",
    agent=recipe_scout
)

review_task = Task(
    description="Review the recipes and rank the top 3 safest options.",
    expected_output="A ranked list of top 3 recipes with clinical justification and caution notes.",
    agent=clinical_reviewer
)
email_task = Task(
    description="Take the final review result, summarize it into a professional email, and send via Gmail.",
    expected_output="Email sent confirmation message.",
    agent=email_agent
)
```

```python
crew = Crew(
    agents=[recipe_scout, clinical_reviewer],
    tasks=[scout_task, review_task],
    verbose=True
)

result = crew.kickoff()
print(result)
```

```python
crew = Crew(
    agents=[recipe_scout, clinical_reviewer, email_agent],
    tasks=[scout_task, review_task, email_task],
    verbose=True
)

# Kickoff Crew
final_result = crew.kickoff()
print(final_result)
```
